```python
class BSTtree:
    """Binary Search Tree implementation with key-value pairs"""
    
    def __init__(self, key, value=None):
        """Initialize a BST node with key and optional value"""
        self.key = key
        self.value = value
        self.left = None
        self.right = None
    
    def insert(self, key, value):
        """
        Insert a key-value pair into the BST
        Returns self to allow method chaining
        """
        if key < self.key:
            # Go to left subtree
            if self.left is None:
                self.left = BSTtree(key, value)
            else:
                self.left.insert(key, value)
        elif key > self.key:
            # Go to right subtree
            if self.right is None:
                self.right = BSTtree(key, value)
            else:
                self.right.insert(key, value)
        else:
            # Key already exists, update the value
            self.value = value
        
        return self  # Return the node for chaining
    
    def find(self, key):
        """
        Find and return the value associated with the key
        Returns None if key doesn't exist
        """
        if key == self.key:
            return self.value  # Return the value, not the node
        elif key < self.key:
            if self.left is None:
                return None
            return self.left.find(key)
        else:  # key > self.key
            if self.right is None:
                return None
            return self.right.find(key)
    
    def update(self, key, value):
        """
        Update the value for an existing key
        If key doesn't exist, insert it
        Returns self to allow method chaining
        """
        if key == self.key:
            self.value = value
        elif key < self.key:
            if self.left is None:
                self.left = BSTtree(key, value)
            else:
                self.left.update(key, value)
        else:  # key > self.key
            if self.right is None:
                self.right = BSTtree(key, value)
            else:
                self.right.update(key, value)
        
        return self  # Return the node for chaining
    
    def delete(self, key):
        """
        Delete a key-value pair from the BST
        Returns the new root of this subtree
        """
        if key < self.key:
            if self.left:
                self.left = self.left.delete(key)
        elif key > self.key:
            if self.right:
                self.right = self.right.delete(key)
        else:
            # Found the node to delete
            # Case 1: No children
            if self.left is None and self.right is None:
                return None
            
            # Case 2: One child
            if self.left is None:
                return self.right
            if self.right is None:
                return self.left
            
            # Case 3: Two children
            # Find the inorder successor (smallest in right subtree)
            successor = self.right._find_min()
            self.key = successor.key
            self.value = successor.value
            # Delete the successor
            self.right = self.right.delete(successor.key)
        
        return self
    
    def _find_min(self):
        """Find the node with minimum key in the subtree"""
        current = self
        while current.left:
            current = current.left
        return current
    
    def _find_max(self):
        """Find the node with maximum key in the subtree"""
        current = self
        while current.right:
            current = current.right
        return current
    
    def inorder_traversal(self):
        """
        In-order traversal: left -> node -> right
        Returns sorted list of (key, value) pairs
        """
        result = []
        if self.left:
            result.extend(self.left.inorder_traversal())
        result.append((self.key, self.value))
        if self.right:
            result.extend(self.right.inorder_traversal())
        return result
    
    def preorder_traversal(self):
        """
        Pre-order traversal: node -> left -> right
        Useful for copying the tree
        """
        result = [(self.key, self.value)]
        if self.left:
            result.extend(self.left.preorder_traversal())
        if self.right:
            result.extend(self.right.preorder_traversal())
        return result
    
    def postorder_traversal(self):
        """
        Post-order traversal: left -> right -> node
        Useful for deleting the tree
        """
        result = []
        if self.left:
            result.extend(self.left.postorder_traversal())
        if self.right:
            result.extend(self.right.postorder_traversal())
        result.append((self.key, self.value))
        return result
    
    def get_height(self):
        """Return the height of the tree (max depth)"""
        left_height = self.left.get_height() if self.left else 0
        right_height = self.right.get_height() if self.right else 0
        return 1 + max(left_height, right_height)
    
    def get_size(self):
        """Return the number of nodes in the tree"""
        left_size = self.left.get_size() if self.left else 0
        right_size = self.right.get_size() if self.right else 0
        return 1 + left_size + right_size
    
    def contains(self, key):
        """Check if a key exists in the BST"""
        return self.find(key) is not None
    
    def get_keys(self):
        """Return all keys in sorted order"""
        return [key for key, _ in self.inorder_traversal()]
    
    def get_values(self):
        """Return all values in key-sorted order"""
        return [value for _, value in self.inorder_traversal()]
    
    def print_tree(self, level=0, prefix="Root: "):
        """Print the tree structure (for debugging)"""
        print("  " * level + prefix + f"({self.key}: {self.value})")
        if self.left:
            self.left.print_tree(level + 1, "L--- ")
        if self.right:
            self.right.print_tree(level + 1, "R--- ")
    
    def is_valid_bst(self, min_key=None, max_key=None):
        """Check if the tree satisfies BST property"""
        if min_key is not None and self.key <= min_key:
            return False
        if max_key is not None and self.key >= max_key:
            return False
        
        left_valid = self.left.is_valid_bst(min_key, self.key) if self.left else True
        right_valid = self.right.is_valid_bst(self.key, max_key) if self.right else True
        
        return left_valid and right_valid


# Example usage and testing
if __name__ == "__main__":
    print("=" * 50)
    print("BINARY SEARCH TREE DEMONSTRATION")
    print("=" * 50)
    
    # Create a BST
    print("\n1. Creating BST with root (50, 'root')")
    bst = BSTtree(50, "root")
    
    # Insert nodes (demonstrating method chaining)
    print("\n2. Inserting nodes...")
    bst.insert(30, "left child")\
       .insert(70, "right child")\
       .insert(20, "left-left")\
       .insert(40, "left-right")\
       .insert(60, "right-left")\
       .insert(80, "right-right")
    print("   Nodes inserted successfully!")
    
    # Display tree structure
    print("\n3. Tree structure:")
    bst.print_tree()
    
    # Test find operations
    print("\n4. Testing find operations:")
    print(f"   Find 30: {bst.find(30)}")
    print(f"   Find 70: {bst.find(70)}")
    print(f"   Find 90: {bst.find(90)}")
    
    # Test update operation
    print("\n5. Testing update operation:")
    bst.update(30, "updated left child")
    print(f"   After update, find 30: {bst.find(30)}")
    
    # Test traversals
    print("\n6. Tree traversals:")
    print(f"   In-order (sorted): {bst.inorder_traversal()}")
    print(f"   Pre-order: {bst.preorder_traversal()}")
    print(f"   Post-order: {bst.postorder_traversal()}")
    
    # Test tree properties
    print("\n7. Tree properties:")
    print(f"   Tree height: {bst.get_height()}")
    print(f"   Tree size (number of nodes): {bst.get_size()}")
    print(f"   All keys (sorted): {bst.get_keys()}")
    print(f"   All values (by key order): {bst.get_values()}")
    print(f"   Contains key 60: {bst.contains(60)}")
    print(f"   Is valid BST: {bst.is_valid_bst()}")
    
    # Test min and max
    print("\n8. Min and max keys:")
    print(f"   Minimum key: {bst._find_min().key}")
    print(f"   Maximum key: {bst._find_max().key}")
    
    # Test deletion
    print("\n9. Testing deletion:")
    print("   Original tree size:", bst.get_size())
    bst.delete(20)
    print("   After deleting 20, size:", bst.get_size())
    print(f"   Does 20 still exist? {bst.contains(20)}")
    
    # Test with different data types
    print("\n10. BST with different data types:")
    bst2 = BSTtree("dog", "canine")
    bst2.insert("cat", "feline")\
         .insert("elephant", "pachyderm")\
         .insert("bird", "avian")
    print(f"    String BST in-order: {bst2.inorder_traversal()}")
    
    print("\n" + "=" * 50)
    print("DEMONSTRATION COMPLETE")
    print("=" * 50)
```
